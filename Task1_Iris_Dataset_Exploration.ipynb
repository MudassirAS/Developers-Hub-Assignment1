```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

Task 1: Exploring and Visualizing a Simple Dataset

Objective: Understand how to read, summarize and visualize a dataset.

Dataset: Iris Dataset

```python
# Load the Iris dataset
iris = sns.load_dataset('iris')

# To see the first 5 rows of the dataset
iris.head()
```

```python
# Check number of rows and columns
print("Shape: ", iris.shape)
```

```python
print("Columns: ", iris.columns)
```

```python
iris.info()
```

```python
# Verify that there are no nulls in the dataset
iris.isnull().sum()
```

```python
# Creating a scatter plot between the sepal_length and sepal_width to see relationship among them for different species

sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title('Sepal Length vs Width')
plt.show()
```

```python
sns.histplot(data=iris, x='petal_length')
plt.title('Distribution of Petal Length')
plt.show()
```

```python
sns.boxplot(data=iris)
plt.title('Box Plot of Iris Features')
plt.xticks(rotation=45)
plt.show()
```

Conclusion:
    - The Iris dataset contains 150 rows and 5 columns (1 for species names, and 4 for features)
    - The dataset is clean and contains no null/missing values
    - Sepal lengthand width vary across species. In terms of width, Setosa is widest, then Versicolor, then Virginica. In terms of length, Virginica is generally the longest, then Versicolor, then Setosa.
    - Petal length distribution is slightly skewed
    - Some outliers are also visible in the box plot
